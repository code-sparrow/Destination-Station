```python
import requests
from pymongo import MongoClient
import pandas as pd

client = MongoClient(port=27017)
db=client.project_2
#station = db.stations.find_one({})
#print(station)

API_KEY = ""
ENDPOINT = "https://api.yelp.com/v3/businesses/search"
HEADERS = {"Authorization": "bearer %s" % API_KEY}
```

```python
museums = []

term = "restaurant"
limit = 3
radius = 805
i = 0
for cursor in db.stations.find({}):
    print(f"{i}---{cursor}")
    print("----------------------------------")
    PARAMETERS = {"term": term,
                "limit": limit,
                "radius": radius,
                "latitude": cursor['start_station_latitude'],
                "longitude": cursor["start_station_longitude"]}


    response = requests.get(url = ENDPOINT, params = PARAMETERS, headers = HEADERS)

    data = response.json()["businesses"]
    #print(data)
    temp = []
    for d in data:
        res = {"name": d["name"], "image_url": d["image_url"], \
                "rating": d["rating"], "display_address": d["location"]["display_address"][0], \
                "display_phone": d["display_phone"], "url": d["url"]}
        temp.append(res)
    museums.append({"station_id": cursor["start_station_id"], "restaurants": temp})
    #print(data)
    print("----------------------------------")
    i += 1


```

```python
restaurants_col = db["restaurants"]
result2 = restaurants_col.insert_many(museums)
```
